Cleaning The data

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("people-1000.xlsx")
print("\n\nCounting is NA Value: ")
print(df.isna().sum())
print("\n\nCounting is NULL Value: ")
print(df.isnull().sum())
```

```python
#Filling Last_name
df['Last Name'].fillna("no last name",inplace=True)
print("\n\nAfter Filling Last Name NA Values: ")
print(df.isna().sum())
```

```python
df['Sex'].fillna(df['Sex'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Email'].fillna('noemail@example.com',inplace=True)
df['Email'].str.strip()
df.isnull().sum()
```

```python
df.dropna()
df.isnull().sum()
```

```python

df['Phone'] = df['Phone'].fillna('000-000-0000')
df['Phone'] = df['Phone'].str.replace('-', '')
df.isnull().sum()
```

```python
df['Phone']
```

```python
df['Job Title'] = df['Job Title'].fillna('Unknown')
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Job Title'] = le.fit_transform(df['Job Title'])
df['Job Title'].value_counts()

```
