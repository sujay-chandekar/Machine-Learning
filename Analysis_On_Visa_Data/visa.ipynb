```python
Q1 = df['prevailing_wage'].quantile(0.25)
Q3 = df['prevailing_wage'].quantile(0.75)
print(Q1,"\n",Q3)
```

```python
IQR = Q3-Q1
IQR
```

```python
Lower = Q1-(1.5*IQR)
print(Lower)
Upper = Q3+(1.5*IQR)
print(Upper)
```

```python
data = df[(df['prevailing_wage']>Upper) | (df['prevailing_wage']<Lower)]
data
```

```python
df['prevailing_wage'] = np.where(
    (df['prevailing_wage'] > Upper) | (df['prevailing_wage'] < Lower),
    df['prevailing_wage'].mean(),
    df['prevailing_wage']
)
```
